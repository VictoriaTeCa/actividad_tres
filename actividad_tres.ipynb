```python
#importamos librer√≠as
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Actividad2_SinNulos.csv')
data
```

```python
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#hacemos copia
data1=data.copy()
```

### **ventas_precios_corrientes**

```python
#diagramas de caja o bigote
#columna por columna para ver si hay datos nulos
fig= plt.figure(figsize=(10,6))
plt.boxplot(data1['ventas_precios_corrientes'])
plt.title('Outliers de ventas_precios_corrientes')
plt.show()
```

```python
#m√©todo cuatiles, encuentro cuartiles de 0.25 y 0.75
y=data1['ventas_precios_corrientes']

percentile25_uno=y.quantile(0.25) #Q1
percentile75_uno=y.quantile(0.75) #Q3
iqr_uno= percentile75_uno - percentile25_uno

Limite_Superior_iqr_uno= percentile75_uno + 1.5*iqr_uno
Limite_Inferior_iqr_uno=percentile25_uno - 1.5*iqr_uno

print('Limite superior permitido', Limite_Superior_iqr_uno)
print('Limite inferior permitido', Limite_Inferior_iqr_uno)
```

```python
#Encontramos Outliers
outliers_iqr_uno= data1[(y>Limite_Superior_iqr_uno)|(y<Limite_Inferior_iqr_uno)]
outliers_iqr_uno
```

```python
data_clean_iqr_uno=data1[(y<=Limite_Superior_iqr_uno)&(y>=Limite_Inferior_iqr_uno)]
data_clean_iqr_uno
```

```python
#diagramas de caja o bigote
fig= plt.figure(figsize=(5,3))
plt.boxplot(data_clean_iqr_uno['ventas_precios_corrientes'])
plt.title('Outliers de Precio venta')
plt.show()

#aparece un valor at√≠pico debido al acomodo de los √≠ndices.
```

```python
data_clean_iqr_uno.to_csv('data_clean_iqr_uno.csv')

from google.colab import files
files.download('data_clean_iqr_uno.csv')
```
